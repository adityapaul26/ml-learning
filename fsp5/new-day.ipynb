```python
# classifcation models
import numpy as np
import pandas as pd


data = pd.read_csv("./Social_Network_Ads.csv") # extracting data from csv
data
```

```python
x = data.iloc[:,1:4].values # extract the values for x
x
```

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder() #encoding to convert strings to numbers

x[:,0]=en.fit_transform(x[:,0])
x
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler() #scaline to imporve accuracy 
x = sc.fit_transform(x)
x
```

```python
y = data.iloc[:,4].values #extracting the values for y
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) #splitting data for test and train

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()

knn_model.fit(x_train,y_train)
```

```python
knn_model.score(x_test,y_test) #checking the accuracy of the model
```

```python
y_pred= knn_model.predict(x_test)
y_pred
```

```python
# to calculate the error we see number of misclassification (might be actual value was 0 but the model predicted 1 pr vice versa)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred) 
cm

# form the confusion matrix we can calculate the accuracy ... it will be equal to the calculated one from before
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy=",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
