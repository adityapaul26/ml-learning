```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,36,49]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r-^") #r means red , g means green...
#p-polygon , c-circle , ^- triangle 
# - for line , -. dotted line
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("x-y dependency")
plt.show()
```

```python
names = ["lightangel" , "bimbok", "bongaon" , "pakhi"]
scores = [8.2 , 7.9, 8.7 , 10.0]
plt.bar(names,scores,color='red',width=.2)
plt.xlabel("names")
plt.ylabel("scores")
#plt.grid() - used to add a grid view
plt.title("student performane")
plt.show()
```

```python
goal_types = ["penalty" ,"free kick", "field"]
goals=[22,12,31]
clr = ["red" , "blue" , "green"]
plt.pie(goals, labels= goal_types,autopct='%.2f%%',explode=(.5,.1,.1),shadow=True,colors=clr)
plt.show()
```
