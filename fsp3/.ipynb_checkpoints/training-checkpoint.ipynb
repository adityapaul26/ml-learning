```python
import numpy as np
import pandas as pd

df = pd.read_csv("./Salary_Data.csv") #read data from csv
df
```

```python
x = df.iloc[:,0:1].values
x
```

```python
y = df.iloc[:,1].values #extract data and convert into numpy array
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)

print("train-data:",x_test)
print("test-data:",x_train)

print("Total size=",x.shape)
print("Train size=",x_train.shape)
print("Test size=",x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
m*2.3+c #salary of someone with experience of 2.3 years
```

```python
model.predict([[2.3]]) #salary of someone with experience of 2.3 years using function
```

```python
model.predict([[1.1]])
#this value exits in the table with a value of 39343
#but we get a diff value from our model ... so there is an error could be +ve or -ve ... 
# we either calculate the mean absolute error or root mean square error (to eliminate the -ve)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_test,y_pred)
print("Mean absolute error=",mae)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("Root mean squared error=",rmse)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,"^")
pr= model.predict(x)
plt.plot(x,pr)
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Salary vs experience")
plt.show()
```
